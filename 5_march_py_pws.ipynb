```python
# Q1
# Import the necessary functions from the bokeh.plotting module:
import bokeh.io
from bokeh.plotting import figure, show
bokeh.io.output_notebook()
#Define two lists containing the data for your line chart:
# prepare some data

x = [1, 2, 3, 4, 5]
y = [6, 7, 2, 4, 5]

#Use the figure() function to create your plot. Pass the following arguments:
#title: the title of your line chart (optional)
# create a new plot with a title and axis labels
p = figure(title="Simple line example", x_axis_label='x', y_axis_label='y')

# add a line renderer with legend and line thickness to the plot
p.line(x, y, legend_label="Temp.", line_width=2)

# show the results
show(p)
```

```python
# Q2
'''A glyph is a vectorized graphical shape or marker that is used to represent your data. A Few Categories of Glyphs: Marker: Shapes like circles, diamonds, squares and triangles.
 Effective for scatter and bubble charts. Line: Single, step and multi-line shapes.'''

from bokeh.io import output_file
from bokeh.plotting import figure, show
bokeh.io.output_notebook()
x = [1, 2, 1]
y = [1, 1, 2]
output_file('first_glyphs.html', title='First Glyphs')
fig = figure(title='My Coordinates',
             plot_height=300, plot_width=300,
             x_range=(0, 3), y_range=(0, 3),
             toolbar_location=None)

fig.circle(x=x, y=y, 
           color='green', size=10, alpha=0.5)
show(fig)
```

```python
# Q3
'''Bokeh comes with five built-in themes: caliber, dark_minimal, light_minimal, night_sky, and contrast.
 Additionally, you can define your own custom themes.
 
 To set the size of your plot, use the attributes width and height when calling the figure() function.
 
 To make your plot automatically adjust to your browser or screen size, use the attribute sizing_mode
 
 can set various attributes to change the way the axes in your plot work and look.'''
from bokeh.io import curdoc
from bokeh.io import output_file
from bokeh.plotting import figure, show
bokeh.io.output_notebook()
x = [1, 2, 1]
y = [1, 1, 2]
# apply theme to current document
curdoc().theme = "dark_minimal"

output_file('first_glyphs.html', title='First Glyphs')
fig = figure(title='My Coordinates',
             plot_height=300, plot_width=300,
             x_range=(0,3), y_range=(0,3),
             toolbar_location=None)

fig.circle(x=x, y=y, 
           color='green', size=10, alpha=0.5)
show(fig)
```
